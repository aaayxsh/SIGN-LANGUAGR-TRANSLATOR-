Image Capture Section

```python
import os
import cv2

# Initialize webcam
cap = cv2.VideoCapture(0)

# Directory to save images
directory = 'Image/'
letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Create directories if they don't exist
for letter in letters:
    letter_dir = os.path.join(directory, letter)
    if not os.path.exists(letter_dir):
        os.makedirs(letter_dir)

while True:
    _, frame = cap.read()

    # Dictionary to keep track of the number of images in each directory
    count = {letter: len(os.listdir(directory + "/" + letter)) for letter in letters}

    # Draw ROI (Region of Interest)
    cv2.rectangle(frame, (0, 40), (300, 400), (255, 255, 255), 2)
    cv2.imshow("Live Feed", frame)
    roi = frame[40:400, 0:300]

    # Handle keypress events to save images
    interrupt = cv2.waitKey(10)
    if interrupt & 0xFF in [ord(letter.lower()) for letter in letters]:
        letter = chr(interrupt & 0xFF).upper()
        cv2.imwrite(os.path.join(directory, letter, f'{count[letter]}.png'), roi)
        print(f"Saved {letter}: {count[letter]}")

    # Exit the loop with 'q'
    if interrupt & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Dataset directory (Update this path to your dataset directory)
dataset_dir = "C:/Users/aaayx/OneDrive/Documents/coding_files/codefiles/projects/realtime sign language translation system/Image"

# Image data generators for training and validation
data_gen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% data for validation
)

# Training data generator
train_gen = data_gen.flow_from_directory(
    dataset_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

# Validation data generator
val_gen = data_gen.flow_from_directory(
    dataset_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Build the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(train_gen.num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,  # Increase epochs for better accuracy
    verbose=1
)

# Save the trained model
model.save("alphabet_recognition_model.h5")
print("Model saved as 'alphabet_recognition_model.h5'")
```

Realtime image prediction 


```python
import cv2
import tensorflow as tf
import numpy as np

# Define letters
letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Load trained model
model = tf.keras.models.load_model("alphabet_recognition_model.h5")

# Start webcam for prediction
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()

    # Region of Interest (ROI)
    cv2.rectangle(frame, (0, 40), (300, 400), (255, 255, 255), 2)
    roi = frame[40:400, 0:300]
    roi_resized = cv2.resize(roi, (64, 64)) / 255.0
    roi_reshaped = np.expand_dims(roi_resized, axis=0)  # Reshape for model input

    # Predict the alphabet
    prediction = model.predict(roi_reshaped)
    predicted_letter = letters[np.argmax(prediction)]
    accuracy = np.max(prediction) * 100

    # Display prediction and accuracy
    cv2.putText(frame, f"Predicted: {predicted_letter} ({accuracy:.2f}%)", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("Prediction", frame)

    # Exit the loop with 'q'
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

Real-Time Prediction and Voice Output

```python
import pyttsx3
import tensorflow as tf
import cv2
import numpy as np

# Define the alphabet letters
letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Load trained model
model = tf.keras.models.load_model("alphabet_recognition_model.h5")

# Initialize text-to-speech engine
engine = pyttsx3.init()

# Start webcam for prediction
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()

    # Draw ROI (Region of Interest)
    cv2.rectangle(frame, (0, 40), (300, 400), (255, 255, 255), 2)
    roi = frame[40:400, 0:300]
    roi_resized = cv2.resize(roi, (64, 64)) / 255.0
    roi_reshaped = np.expand_dims(roi_resized, axis=0)

    # Predict the alphabet
    prediction = model.predict(roi_reshaped)
    predicted_letter = letters[np.argmax(prediction)]
    accuracy = np.max(prediction) * 100

    # Display prediction and accuracy
    cv2.putText(frame, f"Predicted: {predicted_letter} ({accuracy:.2f}%)", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Speak the prediction
    engine.say(f"The letter is {predicted_letter} with {accuracy:.2f} percent confidence")
    engine.runAndWait()

    # Show live feed
    cv2.imshow("Prediction", frame)

    # Exit the loop with 'q'
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
